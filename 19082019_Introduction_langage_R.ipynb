#### 1. Introduction


R est un langage de programmation et un logiciel libre destiné aux statistiques et à la science des données soutenu par la R Foundation for Statistical Computing. R fait partie de la liste des paquets GNU3 et est écrit en C (langage), Fortran et R.

GNU R est un logiciel libre distribué selon les termes de la licence GNU GPL et disponible sous GNU/Linux4, FreeBSD5, NetBSD6, OpenBSD7, Mac OS X8 et Windows9.

Le langage R est largement utilisé par les statisticiens, les data miners, data scientists pour le développement de logiciels statistiques et l'analyse des données.

## Strucure de controle

#### If Then Else

```r
# if then else 

#### On assigne 66 à la variable a

a <- 66 

#### Si a est supérieur à 55 alors
if (a > 55) { 
    #### On affiche "a est supérieur à 55"
    print("a est supérieur à 55") 
} else { 
    #### Sinon 
    print("a est inférieur où égale à 55")
}
```

#### Boucle for

```r
# for loop 
mylist <- c(55, 66, 77, 88, 99) 
#### Pour chaque élement de la liste
for (value in mylist) { 
    #### Afficher l'élement
    print(value) 
}
```

#### Boucle While

```r
# while loop 
a <- 100 

while (a < 500) { 
    a <- a + 100 
} 
a
```

il y a 5 fonctions à maîtriser  que sont :
* mutate() adds new variables that are functions of existing variables

#### Question 1: On calculera la somme totale de nourriture reçu de 2010 à 2015 par l'afghanistan,  haiti et le zimbabwe.

```r
# Read CSV into R
MyData <- read.csv(file="/Users/Coco/Documents/Simplon/TP/tp1_R/Suite_of_Food_Security_Indicators.csv", header=TRUE, sep=",")
```

```r
library(dplyr) #charger la librairie
```

```r
View(MyData) # ne fonctionne pas dans notebook jupiter
```

```r
afgh <- filter(MyData, Area.Code==2) # vue filtrée sur la colonne Area Afghanistan
```

```r
hai <- filter(MyData, Area == 'Haiti') # vue filtrée sur la colonne Area Haiti
```

```r
zim <- filter(MyData, Area == 'Zimbabwe')
```

```r
head(afgh) # affiche les 5 premiers résultats
```

```r
# Création d'une colonne Year.ratio pour pouvoir filter sur l'année effective, affichage uniquement pour le pays Afghaninstan
afgh <- mutate(afgh, Year.ratio =as.numeric(as.character(substr(Year.Code, 1, 4))) + 1) 
hai <- mutate(hai, Year.ratio =as.numeric(as.character(substr(Year.Code, 1, 4))) + 1) 
zim <- mutate(zim, Year.ratio =as.numeric(as.character(substr(Year.Code, 1, 4))) + 1)
```

```r
head(zim)
```

```r
# création d'un filtre sur l'année Year.ratio pour afficher les résultats entre 2010 et 2015
afgh <- filter(afgh, Year.ratio <= 2015 & Year.ratio >= 2010) 
hai <- filter(hai, Year.ratio <= 2015 & Year.ratio >= 2010)
zim <- filter(zim, Year.ratio <= 2015 & Year.ratio >= 2010)
```

```r
zim
```

```r
# afficher la somme selon les différents filtre en amount (year.ratio & pays) de la colonne value
afgh %>% # préciser le datafram qui est exploité 
summarise(afgh_total = sum(as.numeric(as.character(Value))))
hai %>%
summarise(hai_total = sum(as.numeric(as.character(Value))))
zim %>%
summarise(zim_total = sum(as.numeric(as.character(Value))))
```
